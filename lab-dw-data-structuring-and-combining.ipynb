```python
# define function for data cleaning as we did in the previous lab (Cleaning Column Names, 
#Cleaning invalid Values, Formatting data types, Dealing with Null values)

def clean_data(df):
    
    #rename ST and column names in lower case, replace " " by "_"
    df.rename(columns={"ST": "state"}, inplace=True)
    df.columns = df.columns.str.lower().str.replace(" ", "_")
    
    # #Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".
    df["gender"] = df["gender"].str.upper()
    df["gender"] = df["gender"].replace({"female": "F", "Male": "M", "Femal": "F"})

    #State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"
    state_dict = {"AZ": "Arizona", "Cali": "California", "WA": "Washington", "OR": "Oregon", "NE": "Nevada"}
    df['state'] = df['state'].replace(state_dict)
    
    #In education, "Bachelors" could be replaced by "Bachelor"
    df["education"] = df["education"].replace({"Bachelors": "Bachelor"})
    
    #In Customer Lifetime Value, delete the % character and convert Customer lifetime value to numeric
    df["customer_lifetime_value"] = pd.to_numeric(df["customer_lifetime_value"].str.replace("%",""), errors="coerce")
    
    #In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
    luxury_car = {"sports car": "luxury", "luxury suv": "luxury", "luxury car": "luxury"}
    df["vehicle_class"] = df["vehicle_class"].replace(luxury_car)
    
    # Extract the middle value from 'Number of Open Complaints' and convert to numeric
    df["number_of_open_complaints"] = df["number_of_open_complaints"].str.split("/").str[1].astype(float)
    
    # Convert all numeric variables to integers
    numeric_cols = ["customer_lifetime_value", "income", "monthly_premium_auto", "number_of_open_complaints", "total_claim_amount"]
    df[numeric_cols] = df[numeric_cols].astype(int)
    
    return df
```

```python
#cleaning the three files

insurance_company_file1_cleaned_df = clean_data(insurance_company_file1_df)
insurance_company_file2_cleaned_df = clean_data(insurance_company_file2_df)
insurance_company_file3_cleaned_df = clean_data(insurance_company_file3_df)

insurance_company_combined_df = pd.concat([insurance_company_file1_cleaned_df, insurance_company_file2_cleaned_df, insurance_company_file3_cleaned_df], ignore_index=1)
```
